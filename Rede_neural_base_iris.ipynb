```python
#importação do panas
import pandas as pd
```

```python
df = pd.read_csv('/content/IRIS.csv')
#leitura da base de dados, nesse caso, a base de dados iris(encontrada no Kaggle)
```

```python
df
```

```python
df.isnull().sum()
#verificando se há dados nulos
```

TRATAMENTO DA DADOS: DADOS CATEGORICOS PARA VALORES

```python
df.dtypes
#vendo os tipos de dados
```

```python
df['species'] = df['species'].replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})
#trocando(de forma manual) os tipos de dados
```

```python
df.dtypes
```

DEFINIÇÃO DE X E Y

```python
x = df.drop('species', axis=1)
y = df['species']
#definindo as variaveis de analise e alvo
```

CRIAÇÃO DAS VARIAVEIS TESTE E TREINO

```python
from sklearn.model_selection import train_test_split
x_treino,x_teste,y_treino,y_teste = train_test_split(x,y,test_size = 0.3)
#criação do modelo de treinamento
```

```python
x_treino.shape, y_treino.shape
```

```python
x_teste.shape, y_teste.shape
```

CRIAÇÃO DA REDE NEURAL

```python
(3+1)/2
#calculo de quantidades de camadas do neuronio
```

```python
#importação do neuronio e definição dos parametros
from sklearn.neural_network import MLPClassifier
rede_neural_iris = MLPClassifier(max_iter=1500, verbose=True, tol=1e-5, solver='adam', activation='relu', hidden_layer_sizes=(2, 2))
```

```python
rede_neural_iris.fit(x_treino,y_treino)
```

PREVISOES

```python
previsoes = rede_neural_iris.predict(x_teste)
```

```python
acuracia = rede_neural_iris.score(x_teste,y_teste)
print('a acuracia foi de:',acuracia)
#teste de acuracia
```

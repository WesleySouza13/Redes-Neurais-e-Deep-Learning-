```python
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y,test_size = 0.3)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
(34+1)/2
```

```python
rede_neural_Alzheimer = MLPClassifier(max_iter=1500, verbose=True, tol=0.000100, solver ='adam', activation = 'relu', hidden_layer_sizes=(7,7))
rede_neural_Alzheimer.fit(x_treino, y_treino)
```

```python
previsoes = rede_neural_Alzheimer.predict(x_teste)
acuracia = rede_neural_Alzheimer.score(x_teste, y_teste)
print(acuracia)
```
